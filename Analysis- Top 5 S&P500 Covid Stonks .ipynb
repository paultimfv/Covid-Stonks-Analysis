The Covid Pandemic was an interesting time for equities, starting off with heavy selloffs but picking up quickly with the onset of QE before the Fed pivoted to QT and began to raise rates, triggering another market slump. During this 
'pandemic bull market', exactly 21.4 months in which the S&P 500 increased 114.4% on a price basis and 120.4% on a total return basis (price appreciation plus dividends), a number of stocks significantly outperformed the market and generated largely attractive returns for investors. 

What: We will analyze the performances of the top 5 S&P 500 stocks of the pandemic bull market (3/23/2020 - 1/3/2022) with numpy, pandas, matplotlib, and the yfinance libraries. 
How: Calculating daily returns, mean RoR, variance, standard deviation, and correlation 
Stocks: Tesla (+1,281.30%), Caesar's Entertainment(+795.3%), Freeport McMoran(+669.6%), Bath & Body Works(+659.2%), Devon Energy(+637.4%) 
Source: https://www.kiplinger.com/investing/stocks/604798/25-best-sp-500-stocks-of-the-pandemic-bull-market 


Stocks Info: 

Tesla(TSLA) - Elon Musk's Tesla, Inc. designs, develops, manufactures, leases, and sells electric vehicles, and energy generation and storage systems with factories in the United States, China, Netherlands, Germany, and Canada. The firm operates in two segments, Automotive, and Energy Generation and Storage. TSLA surged above 700% in 2020, to the glee of investors and the demise of short-sellers on Wall Street ($38b in shorts wiped). At the time, this surge made Tesla the most valuable car company in the world, rising above established Goliaths like Toyota and Ford. 

Caesars Entertainment(CZR) -  Caesars operates as a US-based gaming and hospitality company which owns, leases, or manages domestic properties in 16 states with slot machines, video lottery terminals and e-tables, poker, as well as hotel rooms. It also operates and conducts sports wagering across 28 jurisdictions in North America. Caesars benefited from a surge in online gambling as more states legalized the service, as casinos closed and Americans hunkered down in their homes, and then even moreso with the "return to normalcy" and the reopening of casinos, as people were excited to return to public places and spend money. 

Freeport McMoran(FCX) - Freeport McMoRan operates in mineral exploration and development; mining and milling of copper, gold, molybdenum and silver; as well as the smelting and refining of copper concentrates, a key commodity used in the construction, manufacturing, and electronics industries. During the pandemic, high demand for Copper largely driven by China brought inventory levels to decade-low levels and Copper prices to their highest levels since 2011. This boosted operating cash flows for Freeport McMoran by as much as +146% at the time and FCX price had performed accordingly. 

Bath & Body Works(BBWI) - Bath & Body Works operates a specialty retailer of home fragrance, body care, soaps, and sanitizer products throughout the US & Canada. During the pandemic, Bath & Body Works' sales boomed from high demand for sanitizer, soap, and body care products, and BBWI prices reflected this strong performance. 

Devon Energy(DVN) - Devon Energy is an independent energy company, which explores for, develops, and produces oil, natural gas, and natural gas liquids in the United States. The aforementioned 'return to normalcy' drove significant demand (and thus prices) for oil and gas. DVN prices in particular rose 164% in 2021, outperforming top oil stocks and generating strong returns for investors. 

```python
import pandas as pd 
import numpy as np 
import yfinance as yf
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
symbols = ['TSLA', 'CZR', 'FCX', 'BBWI', 'DVN']
start_date = '2020-03-23'
end_date = '2022-01-03'

top_covid_stocks = yf.download(symbols, start=start_date, end=end_date)

top_covid_stocks
```

```python
#adjusted closing prices
top_covid_stocks['Adj Close']
```

```python
#a broad overview of stock performance using closing prices

import matplotlib.pyplot as plt

closing_prices = top_covid_stocks['Adj Close']
closing_prices.plot()
plt.xlabel('Date')
plt.ylabel('Adj_Closing_Prices')
plt.title('Top Covid Stocks Quarterly Closing Prices')
```

```python
#broad overview of performance using daily returns 
daily_returns = top_covid_stocks['Adj Close'].pct_change()
daily_returns.plot()
plt.xlabel('Date')
plt.ylabel('Daily_Returns')
plt.title('Top Covid Stocks Daily Returns per Quarter')
```

```python
#to make the above info clearer, we can compare individual returns using subplots. 
#this is a useful comparison of our stocks' volatility 
fig = plt.figure(figsize=(15,15))
ax1 = fig.add_subplot(321)
ax2 = fig.add_subplot(322)
ax3 = fig.add_subplot(323)
ax4 = fig.add_subplot(324)
ax5 = fig.add_subplot(325)
ax1.plot(top_covid_stocks['Adj Close']['TSLA'].pct_change())
ax1.set_title("Tesla")
ax2.plot(top_covid_stocks['Adj Close']['CZR'].pct_change())
ax2.set_title('Caesars Entertainment')
ax3.plot(top_covid_stocks['Adj Close']['FCX'].pct_change())
ax3.set_title("Freeport McMaron") 
ax4.plot(top_covid_stocks['Adj Close']['BBWI'].pct_change())
ax4.set_title("Bed Bath & Beyond")
ax5.plot(top_covid_stocks['Adj Close']['DVN'].pct_change())
ax5.set_title("Devon Energy")
plt.tight_layout()
plt.show()
```

```python
#measuring our mean RoR
import numpy as np 

daily_mean = daily_returns.mean()
daily_mean.keys()
height = []
for key in daily_mean.keys():
    height.append(daily_mean[key])
height
 
x_pos = np.arange(len(daily_mean.keys()))
x_pos

plt.bar(x_pos, height)
plt.xlabel('Top_Covid_Stocks')
plt.xticks(x_pos, daily_mean.keys())
plt.ylabel('daily_mean')
plt.title('Daily Mean Rate of Return')
plt.show()
```

```python
daily_var = daily_returns.var()
daily_var.keys()

height = []
for key in daily_var.keys():
    height.append(daily_var[key])
x_pos = np.arange(len(daily_var.keys()))

plt.bar(x_pos, height)
plt.xlabel('Top_Covid_Stocks')
plt.xticks(x_pos, daily_var.keys())
plt.ylabel('daily_var')
plt.title('Top Covid Stocks Daily Variance')
plt.show()
```

```python
daily_stdev = daily_returns.std()
daily_stdev.keys()

height = []
for key in daily_stdev.keys():
    height.append(daily_stdev[key])
x_pos = np.arange(len(daily_stdev.keys())) 

plt.bar(x_pos, height)
plt.xlabel('Top_Covid_Stocks')
plt.xticks(x_pos, daily_stdev.keys())
plt.ylabel('daily_stdev')
plt.title('Top Covid Stocks Daily Standard Dev.')
```
